```python
import pandas as pd
```

```python
forecast_lightgbm_dic_df = pd.read_csv('./resultados/forecast_lightgbm_dic.csv')
forecast_autogluon_dic_df = pd.read_csv('./resultados/forecast_autogluon_dic.csv')
forecast_autogluon = pd.read_csv('./resultados/autogluon_basico_avanzado.csv')
promedios_mensuales = pd.read_csv('./resultados/promedios_mensuales_productos.csv')
DTW_RL_Exp30_Dic = pd.read_csv('./resultados/DTW_RL_Exp30_Dic.csv')
forecast_regresion_dic_df = pd.read_csv('./resultados/prediccion_regresion2años_exp4_dic.csv')

forecast_lightgbm_feb_df = pd.read_csv('./resultados/forecast_lightgbm_feb.csv')
forecast_autogluon_feb_df = pd.read_csv('./resultados/forecast_autogluon_feb.csv')
forecast_regresion_feb_df = pd.read_csv('./resultados/prediccion_regresion2años_exp4.csv')
```

# Diciembre

```python
forecast_autogluon_dic_df['tn_autogluon'] = forecast_autogluon['Diciembre Autigluon 1']
```

```python
forecast_autogluon_feb_df['tn_autogluon'] = forecast_autogluon['Febrero Autigluon 1']
```

```python
forecast_dic_df = pd.merge(forecast_lightgbm_dic_df, forecast_autogluon_dic_df, on='product_id', how='inner')
forecast_dic_df.drop(columns=['tn'], inplace=True)
forecast_dic_df
```

```python
forecast_dic_df = pd.merge(forecast_dic_df, DTW_RL_Exp30_Dic, on='product_id', how='inner')
forecast_dic_df.rename(columns={'tn': 'DTW_RL_Exp30'}, inplace=True)
forecast_dic_df
```

```python
forecast_dic_df = pd.merge(forecast_dic_df, promedios_mensuales, on='product_id', how='inner')
forecast_dic_df.drop(columns=['Dato real diciembre'], inplace=True)
forecast_dic_df
```

```python
forecast_dic_df.rename(columns={'target_predicted': 'lightgbm'}, inplace=True)
forecast_dic_df.drop(columns=['diff', 'periodo'], inplace=True)
forecast_dic_df
```

```python
forecast_regresion_dic_df
```

```python
forecast_dic_df = pd.merge(forecast_dic_df, forecast_regresion_dic_df, on='product_id', how='inner')
forecast_dic_df.rename(columns={'tn': 'regresion'}, inplace=True)
forecast_dic_df
```

```python
forecast_dic_df['diff_lightgbm'] = abs(forecast_dic_df['lightgbm'] - forecast_dic_df['target'])
forecast_dic_df['diff_autogluon'] = abs(forecast_dic_df['tn_autogluon'] - forecast_dic_df['target'])
forecast_dic_df['diff_DTW_RL_Exp30'] = abs(forecast_dic_df['DTW_RL_Exp30'] - forecast_dic_df['target'])
forecast_dic_df['diff_promedio_201811_201910'] = abs(forecast_dic_df['promedio_201811_201910'] - forecast_dic_df['target'])
forecast_dic_df['diff_regresion'] = abs(forecast_dic_df['regresion'] - forecast_dic_df['target'])
forecast_dic_df
```

```python
# Crear lista de tuplas con (modelo, diferencia) para cada fila
forecast_dic_df['mejor_prediccion'] = forecast_dic_df.apply(lambda x: min([
    #('LIGHTGBM', x['diff_lightgbm']),
    ('AUTOGLUON', x['diff_autogluon']),
    #('DTW_RL_Exp30', x['diff_DTW_RL_Exp30']),
    ('PROMEDIO_201811_201910', x['diff_promedio_201811_201910']),
    ('REGRESION', x['diff_regresion']),
], key=lambda t: t[1]), axis=1)


forecast_dic_df['mejor_modelo'] = forecast_dic_df['mejor_prediccion'].str[0]
forecast_dic_df['mejor_valor'] = forecast_dic_df['mejor_prediccion'].str[1]

forecast_dic_df.drop(columns=['diff_lightgbm', 'diff_autogluon', 'diff_DTW_RL_Exp30', 'diff_promedio_201811_201910', 'diff_regresion', 'mejor_prediccion'], inplace=True)

forecast_dic_df
```

```python
lightgbm_product_ids = forecast_dic_df[forecast_dic_df['mejor_modelo'] == 'LIGHTGBM']['product_id'].unique()
autogluon_product_ids = forecast_dic_df[forecast_dic_df['mejor_modelo'] == 'AUTOGLUON']['product_id'].unique()
promedio_201811_201910_product_ids = forecast_dic_df[forecast_dic_df['mejor_modelo'] == 'PROMEDIO_201811_201910']['product_id'].unique()
DTW_RL_Exp30_product_ids = forecast_dic_df[forecast_dic_df['mejor_modelo'] == 'DTW_RL_Exp30']['product_id'].unique()
regresion_product_ids = forecast_dic_df[forecast_dic_df['mejor_modelo'] == 'REGRESION']['product_id'].unique()
```

# Febrero

```python
forecast_lightgbm_feb_df = forecast_lightgbm_feb_df[forecast_lightgbm_feb_df['product_id'].isin(lightgbm_product_ids)]
forecast_lightgbm_feb_df = forecast_lightgbm_feb_df.groupby('product_id').agg({'target_predicted': 'sum'}).reset_index()
forecast_lightgbm_feb_df.rename(columns={'target_predicted': 'tn'}, inplace=True)
forecast_lightgbm_feb_df
```

```python
forecast_autogluon_feb_df = forecast_autogluon_feb_df[forecast_autogluon_feb_df['product_id'].isin(autogluon_product_ids)]
forecast_autogluon_feb_df = forecast_autogluon_feb_df.groupby('product_id').agg({'tn_autogluon': 'sum'}).reset_index()
forecast_autogluon_feb_df.rename(columns={'tn_autogluon': 'tn'}, inplace=True)
forecast_autogluon_feb_df
```

```python
promedio_201811_201910_df = promedios_mensuales[promedios_mensuales['product_id'].isin(promedio_201811_201910_product_ids)]
promedio_201811_201910_df = promedio_201811_201910_df[['product_id', 'promedio_201811_201910']]
promedio_201811_201910_df.rename(columns={'promedio_201811_201910': 'tn'}, inplace=True)
promedio_201811_201910_df
```

```python
DTW_RL_Exp30_Dic = DTW_RL_Exp30_Dic[DTW_RL_Exp30_Dic['product_id'].isin(DTW_RL_Exp30_product_ids)]
DTW_RL_Exp30_Dic
```

```python
forecast_regresion_feb_df = forecast_regresion_feb_df[forecast_regresion_feb_df['product_id'].isin(regresion_product_ids)]
forecast_regresion_feb_df
```

```python
df = pd.concat([
    forecast_lightgbm_feb_df,
     forecast_autogluon_feb_df, 
     DTW_RL_Exp30_Dic,
     promedio_201811_201910_df,
     forecast_regresion_feb_df
    ], axis=0)
df = df.sort_values(by='product_id', ascending=True)
df
```

```python
df.to_csv('forecast_ensemble.csv', index=False)
```
